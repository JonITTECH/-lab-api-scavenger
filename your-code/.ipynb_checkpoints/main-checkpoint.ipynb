## Challenge 2: Count Commits


As there have been no commits for a long time, instead of searching in the last 3 weeks, I am searching in the last 24 months

```python
import pandas as pd
from datetime import datetime, timedelta
```

```python
import os
import requests
from datetime import datetime, timedelta

os.environ['GITHUB_TOKEN'] = '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'
github_token = os.environ.get('GITHUB_TOKEN')

if not github_token:
    print("Error: The GITHUB_TOKEN environment variable is not set.")
    exit(1)

fecha_limite = datetime.now() - timedelta(weeks=24 * 4)
fecha_limite_iso = fecha_limite.isoformat()

url = f'https://api.github.com/repos/ta-data-pt-rmt/lab-mysql-first-queries/commits?since={fecha_limite_iso}'

headers = {'Authorization': f'token {github_token}'}

def count_commits(url):
    response = requests.get(url, headers=headers)
    if response.status_code == 200:
        commits_data = response.json()
        commit_count = len(commits_data)

        print(f"Total commits in the last 24 months: {commit_count}")

        if 'next' in response.links:
            next_page = response.links['next']['url']
            count_commits(next_page)
    else:
        print(f'Request error. Status code: {response.status_code}')
        print(response.text)

count_commits(url)
```
